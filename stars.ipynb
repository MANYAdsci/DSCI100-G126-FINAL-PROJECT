# Introduction

The most well-known astronomical objects and the primary constituent of galaxies are stars. 
A galaxy's evolution, history and dynamics can be traced by the composition of its stars. Astronomers started classifying stars based on their mass and temperature. Astronomer Annie Jump Cannon devised Seven categories (also called, classes) to classify stars. The classes are called O, B, A, F, G, K and M. It is on the basis of their spectra (the elements that they absorb) and their temperature. ("Oh be a fine guy/girl, kiss me" is a simple mnemonic to remember the classes.)

# Question

Which is the brightest star among all of the classes of stars?

To answer this question we will analyze a data set of stars from NASA. This data was downloaded from the website Kaggle (https://www.kaggle.com/datasets/brsdincer/star-type-classification) & compiled into a single CSV. Each row in the data set contains the values of categories a star is classified into. For our data set, we will be taking four predictors into account: temperature, relative luminosity, relative radius and absolute magnitude. 
For our analysis, we are going to use the stars dataset.

To analyze this, we will be looking at a dataset that includes the following columns:



Temperature: Temperature in Kelvin (K) 

L --L/Lo (Relative Luminosity): It is the total brightness of a star. 

R -- R/Ro (Relative Radius): It is the relative radius of a star.

A_M -- Mv (Absolute Magnitude): It is the magnitude a star would have if it were viewed at a standard distance of 10 parsecs (32.6 light-years). 

Color (General Obs. Color): The general color of the spectrum of the star.

Spectral_Class (SMASS Spectral): The spectrum emitted by the star. 

Type (Type-One from 0-5): 

Based on its position with the scale, stars can be subdivided withing each stellar type. We have limited our dataset to 6 subtypes of stars and assigned them the following numbers from 0 to 5:

o   Red Dwarf - 0 

o   Brown Dwarf - 1 

o   White Dwarf - 2 

o   Main Sequence - 3

o   Super Giants - 4 

o   Hyper Giants - 5 

```r
library(readxl)
library(tidyverse)
library(repr)
library(tidymodels)
```

```r
stars<- read.csv("data/Stars.csv")
stars<- mutate(stars, Spectral_Class = as.factor(Spectral_Class))
stars
```

```r
set.seed(200)

stars_recipe <- recipe(Spectral_Class ~ Temperature + L + R + A_M, data = stars) %>%
    step_scale(all_predictors()) %>%
    step_center(all_predictors())

stars_scaled<- stars_recipe %>%
    prep()%>%
    bake(stars)

stars_scaled %>%
select(A_M, L, Spectral_Class)
```

```r
stars_split <- initial_split(stars_scaled, prop= 0.75, strata= Spectral_Class)
stars_train <- training(stars_split)
stars_test <- testing(stars_split)
```

```r
stars_plot <- ggplot(stars_train, aes(x= L, y= A_M, colour= Spectral_Class))+
geom_point()+
xlab("Relative Luminosity L/Lo (scaled)")+
ylab("Absolute Magnitude (Mv) (scaled)") +
theme(text= element_text (size=24))
stars_plot

#Visualization of training data of star spectral type based on predictors Temperature and Luminosity
```

Most M stars have high temperature and low luminosity. Most O stars have low 
temperature and a range of luminosity. Most B, F, G stars have average temperature and 
low luminosity.
